```python
import random
from faker import Faker

fake = Faker()
```

```python
# Number of records to generate
# Insert statements for relation Person is provided
# Complete generating insert statements for all other relations based on the counts below
num_person = 100
num_faculty = 10
num_student = 90
num_grad_student = 15
num_committee = 5
num_grant_info = 3
num_instructor_researcher = 25
num_support = 5
num_college = 3
num_department = 5
num_course = 7
num_section = 0
num_current_section = 7
num_transcript = 0
num_registered = 90
num_majorminor = 105
num_belongs = 10

# Generate PERSON records
print("-- INSERT INTO PERSON Table")
for personid in range(num_person):
    fname = fake.first_name()
    minit = fake.random_letter().upper()
    lname = fake.last_name()
    bdate = fake.date_of_birth(minimum_age=18, maximum_age=70)
    gender = random.choice(['M', 'F'])
    street = fake.street_address()
    apt_no = random.randint(100, 999)
    city = fake.city()
    state = fake.state_abbr()
    zip_code = fake.zipcode()
    
    print(f"INSERT INTO PERSON (Fname, Minit, Lname, Bdate, gender, Street, Apt_no, City, State, Zip) "
          f"VALUES ('{fname}', '{minit}', '{lname}', '{bdate}', '{gender}', '{street}', '{apt_no}', '{city}', '{state}', '{zip_code}');")
```
